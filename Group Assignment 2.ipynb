Group 20’s Airport idea required an ER Diagram for the project. With this, it was our group’s intention to provide an easy-to-read diagram in order to fulfill the requirements.

Our first entity that we will mention is the Passenger Ticket. Passenger Ticket is connected to all of our other entities due to its relevance to other entities. It’s primary key nickname is “Ticket Number”. It contains attributes such as: class, plane, destination, flight number, gate, Passenger ID number, and round trip. Passenger Ticket’s foreign keys include plane, which connects it to plane due to its flight assignment, Passenger ID Number, which connects it to the Passenger’s information. Its last foreign key is round trip, which allows us to tell whether a passenger is returning to our airport. 

Our second entity is Flights. The Flights entities contains the following attributes: plane, pilot, Name of Destination, time of departure, time of arrival, and Gate Number, with Gate Number and Name of Destination being the foreign keys. It is connected to Passenger Ticket because every ticket is assigned to a single flight route. It is also connected to Destination because each flight must go somewhere. The final connection is Plane because each ticket must be assigned to a single Plane. Flights’ primary key is “flight number”. 

Our third entity is Plane. This entity contains no foreign keys but its Primary key is “plane ID Number”. Its attributes are size, model, condition, carry on, and seats (which connects to the Seats entity). It has relationships to both seats and flights as well as Passenger Ticket. Each plane must have seats and must participate in a flight. It must also have passengers which have their Passenger Tickets. 

The Seats entity is our fourth object. Its primary key is “plane seat ID”. It contains three categories for seats: coach, business class, and first class. This gives our customers three different options that help suit their needs during their flights. This entity contains no foreign keys. 

For our fifth entity, we included a table for Destination. With its primary key being “Name of Destination”, the Destination entity allows us to organize what flights are going where and which passengers are going to what places. We have included attributes such as: distance from airport, hemisphere, passport required, and destination airport. These attributes help us organize the final destinations that our passengers would like to be dropped off. 

One of our most crucial entities is Gate. Our intention was to reduce flight delays, and this entity helps us with doing so. With this entity, we can separate different flights such as international, domestic, and connecting flights. This will provide us with special data that is separated into different categories. 

Delays is another crucial entity to our design. It contains the attributes original time, rescheduled time, total time delayed, and flight number. This entity is debatably the most important to our project and its intention is to provide data on flight delays. This goes in tandem with our Gate entity as well as the Flight entity as well. 

Our last entity is Passenger Contact Info. It contains the primary key “Passenger ID number” and lists phone number, address, and log-in info as its attributes. This allows us to contact our customers in case their flight is delayed or if we can provide them a discount or something along those lines. Phone number and address are crucial for contacting our customers, which is why we included them. We also have them in case they forget their log-in information as well. 

Our primary entity is Passenger Ticket due to its relevance to each of the other entities that we have provided. Our forgein keys include: plane, Passenger ID number, round trip, flight number, Name of Destination, and Gate Number.
